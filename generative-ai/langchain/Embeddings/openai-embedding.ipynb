### OpenAI Embedding Techniques

```python
# Langchain imports
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Community integrations
from langchain_community.vectorstores import Chroma
from langchain_community.document_loaders import TextLoader

# OpenAI intergrations
from langchain_openai import OpenAIEmbeddings

# Environment variables
from dotenv import load_dotenv

load_dotenv() 
```

```python
# Embedding the text using openai embedding model
text = "Hi this is the testing of openai embedding models for converting text to vectors."
embeddings = OpenAIEmbeddings(model='text-embedding-3-large')
embedded_text = embeddings.embed_query(text)
embedded_text
```

```python
# Checking the vector dimensions
len(embedded_text)
```

```python
# Embedding the text using openai embedding model with reduced vector dimensions
text = "Hi this is the testing of openai embedding models for converting text to vectors."
embeddings_1024 = OpenAIEmbeddings(model='text-embedding-3-large', dimensions=1024)
embedded_text = embeddings_1024.embed_query(text)
embedded_text
```

```python
# New vector dimensions after reduction
len(embedded_text) 
```

#### Workflow Example from Data Ingestion to Embeddings

1. Data Ingestion (Loading Text from Data Source)
2. Data Transformer (Splitting Text into Text Chunks)
3. Embeddings (Text Chunks to Vectors)


```python
# Loading the text documents from a flat file
text_loader = TextLoader('speech.txt')
text_documents = text_loader.load()
text_documents
```

```python
# Splitting the text documents into smaller chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
splitted_documents = text_splitter.split_documents(text_documents)
splitted_documents
```

```python
# Embedding the splitted documents into vectors
embeddings = OpenAIEmbeddings(model='text-embedding-3-large')
embedded_documents = embeddings.embed_documents([doc.page_content for doc in splitted_documents])
embedded_documents
```

```python
# Combined Embedding conversion and storing it in the vector store
vector_store = Chroma.from_documents(embedded_documents, embeddings)
```

```python
# Retrieving results from querying the vector store
query = "It will be all the easier for us to conduct ourselves as belligerents"
retrieved_results = vector_store.similarity_search(query)
retrieved_results
```
