#### How to split by character using CharacterTextSplitter

```python
# Langchain import
from langchain_text_splitters import CharacterTextSplitter

# Community integration
from langchain_community.document_loaders import TextLoader
```

```python
# Loading text documents from a flat file
text_loader = TextLoader('speech.txt')
text_documents = text_loader.load()
text_documents
```

```python
# Splitting the text documents using CharacterTextSplitter
text_splitter = CharacterTextSplitter(separator="\n\n", chunk_size=1000, chunk_overlap=50)
splitted_documents = text_splitter.split_documents(text_documents)
splitted_documents
```

```python
# Loading the text from the flat file
with open('speech.txt', 'r') as file:
    text = file.read()

text
```

```python
# Splitting the text using CharacterTextSplitter
text_splitter = CharacterTextSplitter(separator="\n\n", chunk_size=1000, chunk_overlap=50)
splitted_text = text_splitter.create_documents([text])
splitted_text
```
