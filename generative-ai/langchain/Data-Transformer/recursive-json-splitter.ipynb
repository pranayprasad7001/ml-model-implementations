#### Data Transformer

##### How to Split JSON Data using RecursiveJsonSplitter

```python
import requests
from langchain_text_splitters import RecursiveJsonSplitter
```

```python
# Loading json data from a url
json_data = requests.get("https://raw.githubusercontent.com/mledoze/countries/master/countries.json").json()
json_data
```

```python
# Splitting the json data into smaller chunks
json_splitter = RecursiveJsonSplitter(max_chunk_size=500)
json_chunks = json_splitter.split_json(json_data, convert_lists=True)
json_chunks
```

```python
for chunk in json_chunks[:3]:
    print(chunk)
```

```python
# Splitter can also output document types
docs = json_splitter.create_documents(texts=[json_data], convert_lists=True)
docs
```
