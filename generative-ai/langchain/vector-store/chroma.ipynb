### VectorStoreDB - Chroma

```python
# Langchain imports
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Community integrations
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import OllamaEmbeddings

# Chroma integrations
from langchain_chroma import Chroma
```

```python
# Loading the documents from the flat file
loader = TextLoader('speech.txt')
text_documents = loader.load()
text_documents
```

```python
# Splitting the documents into smaller chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=0)
splitted_text_documents = text_splitter.split_documents(text_documents)
splitted_text_documents
```

```python
# Initializing the ollama embedding model
embeddings = OllamaEmbeddings(model='qwen3-embedding:4b')
embeddings
```

```python
# Embedding chunks into vectors and storing it in the vectore store 
vectore_store_1 = Chroma.from_documents(splitted_text_documents, embeddings)
vectore_store_1
```

```python
# Querying the vector store using similarity_search 
query_text = "What does the speaker believe is the main reason the United States should enter the war?"
query_results = vectore_store_1.similarity_search(query_text)
query_results
```

```python
query_results[0].page_content
```

```python
# Storing the vector store locally
vectore_store_2 = Chroma.from_documents(splitted_text_documents, embeddings, persist_directory="./chroma_db")
```

```python
# Loading the vector store from the disk
vectore_store = Chroma(persist_directory="./chroma_db", embedding_function=embeddings)
query_results = vectore_store_2.similarity_search(query_text)
query_results[0].page_content
```

```python
# Querying the vector store using similarity_search_with_score 
query_results_with_score = vectore_store_2.similarity_search_with_score(query_text)
query_results_with_score
```

```python
# Converting the vector store into retriever and querying by invoking it
retriever = vectore_store_2.as_retriever()
query_results = retriever.invoke(query_text)
query_results[0].page_content
```
