#### Simple LLM LCEL

```python
# Langchain imports
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate

# Groq integrations
from langchain_groq import ChatGroq

# Environment variables
from dotenv import load_dotenv

load_dotenv()
```

```python
# Initializing the LLama Model
model = ChatGroq(model="llama-3.1-8b-instant")
```

```python
# Creating Messages for the Model without using Prompt Template and its Response 
messages = [
    SystemMessage(content="Translate from English to French"),
    HumanMessage(content="Hello, How are you?")
]

response = model.invoke(messages)
response
```

```python
# Initializing and Parsing the Model Response
parser = StrOutputParser()
result = parser.invoke(response)
result
```

```python
# Chaining components using LCEL
chain = model | parser
chain.invoke(messages)
```

```python
# Creating Prompt Template
generic_template = "Translate the following into {language}:"
prompt = ChatPromptTemplate.from_messages(
    [
        ("system", generic_template), 
        ("user", "{text}")      
    ]
)
```

```python
# Model Output when using Prompt Template
result = prompt.invoke({"language": "French", "text": "Hello, How are you?"})
result
```

```python
result.to_messages
```

```python
# Chaining all the Components Together
chain = prompt|model|parser
chain.invoke({"language": "French", "text": "Hello, How are you?"})
```
