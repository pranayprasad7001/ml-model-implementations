### OpenAI ChatBot

```python
# LangChain imports
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# OpenAI integrations
from langchain_openai import ChatOpenAI

# Environment variables
from dotenv import load_dotenv

load_dotenv()
```

```python
# Initializing the openai llm model
llm = ChatOpenAI(model='gpt-4o')
print(llm)
```

```python
# Querying the llm by invoking it with the input
response_result = llm.invoke("What is Generative AI?")
response_result
```

```python
# Creating the prompt template 
prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are an expert AI Engineer. Provide me answers based on the question"),
        ("user", "{input}")
    ]
)

prompt
```

```python
# Chaining the prompt template with the llm
chain = prompt | llm
response = chain.invoke({"input": "Can you tell me about Langsmith?"})
response
```

```python
# Initializing the output parser 
output_parser = StrOutputParser()
```

```python
# Creating the chain and querying by invoking it
chain = prompt | llm | output_parser
chain.invoke({'input': "Can you tell me about Langsmith?"})
```
